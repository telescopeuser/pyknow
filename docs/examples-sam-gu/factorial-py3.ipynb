Page 5 - Computing the factorial

```python
from pyknow import *
#watch()
```

```python
class Factorial(Fact):
    pass
```

```python
def isint(x):
    return isinstance(x, int)
    
class ComputeFactorial(KnowledgeEngine):    
    @Rule(Factorial('x' << P(isint),
                    'y' << P(isint)))
    def factorial(self, x, y):
        self.declare(Factorial(x + 1, (x + 1) * y))
```

```python
cf = ComputeFactorial()
cf.reset()
cf.declare(Factorial(1, 1))
cf.run(35)
cf.facts
```
